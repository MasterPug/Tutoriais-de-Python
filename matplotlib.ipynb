## Matplotlib <br>

### [Funções básicas]
>Vamos começar importando um módulo da matplotlib chamada pyplot, ela provê funções simples para a adição de elementos,
como linhas, imagens e textos.

```python
import matplotlib.pyplot as gr
import numpy as np
```

```python
gr.plot([1,2,3,4],[25,98,45,4])
gr.show()
```

Na função `plot()` eu passei dois arrays aos quais são o eixo x e y sucessivamente, e com o `show()` eu mostrei o gráfico gerado, lembrando que o `gr` é apenas minha váriavel que pode assumir qualquer nome.

E dentro dessa lógica podemos alterar alguns detalhes visuais desse gráfico, como por exemplo adicionar um título a esse plot, nomear os eixos, vou demonstrar abaixo com as funções `title()`, `xlabel()`, `ylabel()`,`suptitle()`.

```python
gr.plot([1,2,3,4],[25,98,45,4])
gr.title('eu sou o titulo')
gr.xlabel('eu sou o x')
gr.ylabel('eu sou o y')
gr.suptitle('eu sou o subtitulo')
gr.show()
```

Você também pode especificar o tamanho da figura que será mostrado atráves do `figure(figuresize=(numero de linhas, numero de colunas))`, mostrar as grades do gráfico atráves do método `grid()`, e também dentro da função plot podemos adicionar outras configurações(ex: colocar mais uma linha, trocar a cor da linha, desativar a linha e etc). Definindo labels para cada plot, podemos chamar o método `legend()`, ele cria legenda para o gráfico. Você pode limitar os valores que aparecem no eixo x e y atráves do `xlim ou ylim`.

```python
gr.figure(figsize=(10,5))
gr.title('eu sou o titulo')
gr.xlabel('eu sou o x')
gr.ylabel('eu sou o y')
gr.suptitle('eu sou o subtitulo')
gr.xlim(1.5,9)
gr.ylim(10,100)
gr.plot([1,2,3,4],[25,98,45,4],"ro", label='primeira')
gr.plot([2,5,6,9],[22,23,31,31],"g^", label='segunda')
gr.grid()
gr.legend()
gr.show()
```

Você pode fazer plots separados atráves do `subplot()`, os parâmetros que você passa é o número de linhas e em seguida a coluna, e por último o index do subplot.

```python
##gr.figure(figsize=(10,5))
gr.subplot(1,2,1)
gr.title('eu sou o titulo')
gr.xlabel('eu sou o x')
gr.ylabel('eu sou o y')
gr.plot([1,2,3,4],[25,98,45,4],"ro")
gr.grid()

gr.suptitle('plots separados')


gr.subplot(1,2,2)
x = np.arange(0,10)
data = 3*x
gr.plot(data, "go")
gr.title('eu sou o subtitulo 2')

gr.show()
```

### [Gráfico de Barras]

- Você também pode representar seus dados atráves de um gráfico de barras

Você passa um array com os nomes e outro com os valore, para o método `bar()`, ao qual faz o plot do gráfico de barras e você pode definir uma cor.

```python
nomes = ["dado1", "dado2", "dado3", "dado4", "dado5"]
valores = [31,65,88,21,1]
gr.bar(nomes,valores,color='purple')
gr.title("agora é barras")
gr.xlabel('nome de cada parâmetro')
gr.ylabel('valores dos dados')
gr.show()
```

Quando você desejar fazer um comparativo é bem interessante colocar dois gráficos lado a lado, para uma melhor visualização. Neste caso, você vai precisar passar a posição das barras, que no caso do exemplo eu defini como `index`, e irá passar passar como o primeiro parâmetro do `bar()` logo depois os valores e em seguida a largura da barra. Neste caso torna-se interessante também mostrar o método `xticks()` ao qual você passa a posição em que os nomes dos parâmetros ficarão, e também podemos ver o `legend()` recebendo a posição que a legenda ficará.

```python
nomes = ["dado1", "dado2", "dado3", "dado4", "dado5"]
valores = [31,65,88,21,1]
valores2 = [35,79,120,31,2]

index = np.arange(5)
largura = 0.25

gr.bar(index,valores,largura,color='purple', label='valores1')
gr.bar(index+largura, valores2, largura, color='green', label="valores2")
gr.title("agora é barras")
gr.xlabel('nome de cada parâmetro')
gr.ylabel('valores dos dados')
gr.xticks(index+largura/2,nomes)
gr.legend(loc='upper left')
gr.show()
```

Podemos também fazer gráficos de barra horizontalmente atráves do `barh()`

```python
nomes = ["dado1", "dado2", "dado3", "dado4", "dado5"]
valores = [31,65,88,21,1]
variancia = [3,2,8,9,1]

gr.barh(nomes,valores,xerr=variancia, color='purple')
gr.title("agora é barras")
gr.xlabel('nome de cada parâmetro')
gr.ylabel('valores dos dados')
gr.show()
```

Podemos também colocar um acima do outro, atráves do parâmetro `bottom`.

```python
nomes = ["dado1", "dado2", "dado3", "dado4", "dado5"]
valores = [31,65,88,21,1]
valores2 = [35,79,120,31,2]

index = np.arange(5)
largura = 0.25

gr.bar(index,valores,largura,color='purple', label='valores1')
gr.bar(index, valores2, largura, color='green', label="valores2", bottom=valores)
gr.title("agora é barras")
gr.xlabel('nome de cada parâmetro')
gr.ylabel('valores dos dados')
gr.xticks(index,nomes)
gr.legend(loc='upper left')
gr.show()
```

### [Histograma]

Temos também a opção de fazer um histograma para a representação dos dados.

```python
x = np.random.randn(1000)
gr.title("Agora é histograma")
gr.xlabel("Esse é o x")
gr.ylabel("Esse é o y")
gr.hist(x,10,color="green")
gr.show()
```

### [Gráfico de pizza]

Também temos o gráfico de pizza, ou chamado de pie.

```python
nomes = ["dado1", "dado2", "dado3", "dado4", "dado5"]
valores = [31,65,88,21,1]
Explode = [0,0.1,0,0,0]
gr.pie(valores, explode=Explode,labels = nomes, shadow=True, startangle=25)
gr.axis('equal')
gr.legend(title='Lista de dados',loc="lower right")
gr.show()
```

```python
vet1 = np.array([80,50,42,83,62,86,18,57,72,4,27,34,79,22,30,21,39,63,63,84])
vet2 = np.array([38,95,55,87,91,75,49,77,74,47,64,75,61,13,93,68,53,83,73,64])

gr.xlim(1,100)
gr.ylim(1,95)
gr.scatter(vet1, vet2, color="purple")
gr.xlabel("Esse é o x")
gr.ylabel("Esse é o y")
gr.show()
```
